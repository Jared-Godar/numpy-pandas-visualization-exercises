### Crosstabs vs. Pivot tables
- Crosstabs are summations of categoryA by categoryB
    - Row is each value of categoryA
    - Column is each value of category B


```python
df.head()
```

```python
pd.crosstab(df.manufacturer, df['class'])
```

- Pivot tables are any kind of aggrecation of valueC by categoryA by categoryB
    - default to average

```python
#df.pivot_table(df, index='CategoryA', columns='categoryB', values='numericColumns')
pd.pivot_table(df, index='manufacturer', columns='class')
```

```python
#Defaults to average of the calues argument/column
#NaNs - dont make that car

pd.pivot_table(df, index='manufacturer', columns='class', values='cty')
```

- Cross tab is just a count
- Pivot defaults to average of pairwise combinations
