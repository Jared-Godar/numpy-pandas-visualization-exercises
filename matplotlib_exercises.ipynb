4. <p>Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.</p>

```python
plt.figure(figsize = (10,8))

x2a = range(0, 4)
y2a = [(math.sqrt(x2a)) for x2a in x2a]

x2b = range(-3, 4)
y2b = [x**3 for x in x2b]

x2c = range(0, 4)
y2c = [2**x for x in x2c]

x2d = range(0, 4)
y2d = [(1/(x+1)) for x in x2d]

plt.plot(x2a, y2a,
         'k--*', 
         label = '$sqrt(x)$') 

plt.plot(x2b, y2b,
         color = 'blue',
         linestyle = '-', 
         marker = '',
         label = '$x^3$') 

plt.plot(x2b, y2b,
         color = 'blue',
         linestyle = '-', 
         marker = '',
         label = '$x^3$') 

plt.plot(x2c, y2c,
         color = 'green',
         linestyle = '-', 
         marker = '',
         label = '$2^x$') 

plt.plot(x2d, y2d,
         color = 'red',
         linestyle = '-', 
         marker = '',
         label = '1/((x+1))') 

plt.title('Exercise 4 Combined Graph')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')


plt.legend()
plt.show()
```
